```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Wisconsin breast cancer data set, which is available as part of the sklearn datasets.
# You can read more about this dataset here: https://scikit-learn.org/stable/datasets/toy_dataset.html#breast-cancer-dataset
cancer = datasets.load_breast_cancer()
```

```python
# The object cancer is a dictionary, you can check the keys using
cancer.keys()
```

```python
# and check the details by calling the keys, like by using
cancer.feature_names
```

```python
# to see the feature names. You can also view a more detailed description of the dataset
print(cancer.DESCR)
```

```python
# Please read the description! On your own, explore the other functions related to the keys. You
# will also want to load the data (from cancer.data) into a pandas DataFrame.
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['label'] = cancer.target
df
```

```python
# Question 1:
df

# Question 2:
malignant = sum(cancer.target == 0)
malignant

# Question 3:
plt.xlabel('Mean Perimeter')
plt.ylabel('Mean Compactness')
plt.title('Mean Compactness vs. Mean Perimeter')
plt.scatter(df[df['label'] == 0]['mean perimeter'], df[df['label'] == 0]['mean compactness'], 
            color='tab:blue', label='Malignant', alpha=0.75, s=50)
plt.scatter(df[df['label'] == 1]['mean perimeter'], df[df['label'] == 1]['mean compactness'], 
            color='tab:red', label='Benign', alpha=0.75, s=50)
plt.legend()
plt.show()

# Question 4:
data = df[['mean texture', 'mean compactness', 'mean radius', 'mean perimeter']].copy()
data['diagnosis'] = cancer.target
data['diagnosis'] = data['diagnosis'].map({0: 'malignant', 1: 'benign'})
sns.pairplot(data, hue='diagnosis', diag_kind='kde', palette='magma')
plt.show()
```
